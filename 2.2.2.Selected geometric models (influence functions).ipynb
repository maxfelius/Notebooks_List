```python
import numpy as np
import os, sys, time
import matplotlib.pyplot as plt
```

```python
def zg(R,r):
    return -(1/R**2)*np.exp(-np.pi*(r**2/R**2))

def func1(R,r):
    return -(3/(np.pi*R**2))*(1-(r/R)**2)**2

def func2(R,r):
    return -(2/(np.pi*np.sqrt(np.pi)*R))*(1/r)*np.exp(-4*(r/R)**2)

def func3(R,r):
    limit = np.deg2rad(35)
    H = R/np.tan(limit)
    zeta = np.arctan(r/H)
    return -np.cos(zeta)**2

def func4(R,r):
    return -2.256*(1/r)*np.exp(-4*r**2)

def func5(R,r):
    return -0.1392*np.exp(-0.5*r**2)
```

```python
x = np.arange(-20,20,0.1)
R = 7

z1 = zg(R,x)
z2 = func1(R,x)
z3 = func2(R,x)
z4 = func3(R,x)
z5 = func4(R,x)
z6 = func5(R,x)
```

```python
plt.figure(figsize=(10,7))
plt.plot(x,-z1/np.min(z1),label='Eq.(2.5)')
plt.plot(x,-z2/np.min(z2),label='Eq.(2.6)')
plt.plot(x,-z3/np.min(z3),label='Eq.(2.7)')
plt.plot(x,-z4/np.min(z4),label='Eq.(2.8)')
plt.plot(x,-z5/np.min(z5),label='Eq.(2.9)')
plt.plot(x,-z6/np.min(z6),label='Eq.(2.10)')

plt.grid(True)
plt.ylabel('Normalized $k_z$ [m]',size=14)
plt.xlabel('Radial distance [m]',size=14)
plt.legend();
```
