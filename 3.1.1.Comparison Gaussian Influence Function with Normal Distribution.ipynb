# Comparison Gaussian Influence Function with Normal Distribution
@author: Max Felius

\begin{equation}
    g(x) = \frac{1}{\sigma \sqrt{2\pi}}\exp\left(\frac{-(x - \mu)^2}{2\sigma^2} \right)
\end{equation}

```python
import numpy as np
import os, sys, time
import matplotlib.pyplot as plt

def zg(R,r):
    return -(1/R**2)*np.exp(-np.pi*(r**2/R**2))

def gauss(R,r):
    sigma = R
    x = r
    mu = 0
    return (1/(sigma*np.sqrt(2*np.pi)))*np.exp(-0.5*((x-mu)**2)/(sigma**2))

def gauss2(R,r):
    sigma = R
    x = r
    mu = 0
    return np.exp(-0.5*((x-mu)**2)/(sigma**2))
```

```python
x = np.arange(-20,20,0.1)
R = 7
z1 = zg(R,x)

fig,ax = plt.subplots(figsize=(10,7))

ax.annotate(s='', xy=(R,0), xytext=(0,0), arrowprops=dict(arrowstyle='->'),fontsize='xx-large')
plt.text(R/3,0.01,'R',fontsize='xx-large')

ax.plot(x,R**2 * z1,label='Gaussian Influence Function')
ax.scatter(R,0,c='k')
ax.scatter(0,0,c='k')

ax.plot(x,-gauss2(R,x),label='Normalized Normal Distribution')
ax.plot([R,R],[0,-1],c='k',linestyle='--')

ax.grid(True)
ax.legend(fontsize=14)
ax.set_ylim([-1,0.05])
ax.set_xlabel('Radial Distance [m]',size=14)
ax.set_ylabel('Normalized Subsidence [m]',size=14);
```
